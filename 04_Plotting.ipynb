## X,Y Plots

```python
d = np.linspace(1,10,200)
I = np.exp(-d)
plot(d,I)
```

```python
d = np.linspace(1,10,200)
I = np.exp(-d)
plot(d,I)
ylabel('I')
xlabel('d')
```

```python
d = np.linspace(1,10,200)
I = np.exp(-d)
plot(d,I)
ylabel(r'$I= \exp(-d)$')
xlabel('d')
```

```python
d = np.linspace(1,10,200)
I = d**(-2)
plot(d,I)
ylabel(r'$I= 1 / d^2$')
xlabel('d')
```

```python
d = np.linspace(1,10,200)
Iid2 = d**(-2)
plot(d,Iid2)
ylabel(r'$I$')
Inex = np.exp(-d)
plot(d,Inex)
xlabel('d')
```

```python
d = np.linspace(1,10,200)
Iid2 = d**(-2)
plot(d,Iid2, label = r'$I = d^{-2}$')
ylabel(r'$I$')
Inex = np.exp(-d)
plot(d,Inex, label = r'$I = \exp(-d)$')
xlabel('d')
legend()
```

```python
d = np.linspace(1,10,200)
Iid2 = d**(-2)
plot(d,Iid2, 'r-', label = r'$I = d^{-2}$')
ylabel(r'$I$')
Inex = np.exp(-d)
plot(d,Inex, 'g--',label = r'$I = \exp(-d)$')
xlabel('d')
legend()
```

```python
d = np.linspace(1,10,200)
Iid = d**(-1)
plot(d,Iid, 'b-.', label = r'$I = d^{-1}$')
Iid2 = d**(-2)
plot(d,Iid2, 'r-', label = r'$I = d^{-2}$')
ylabel(r'$I$')
Inex = np.exp(-d)
plot(d,Inex, 'g--',label = r'$I = \exp(-d)$')
xlabel('d')
legend()
title('Distance Decay')
```

## Scatter Plot

```python
x = np.arange(100)
y = 0.5 + np.random.normal(size=(100,))
scatter(x,y)
xlabel('x')
ylabel('y')
```

```python
scatter(x,y)
vlines(x.mean(),y.min(),y.max())
hlines(x.min(),x.max(),y.mean())
```
